# **Ejemplos 1**
En este ejemplo crearemos dos sorting algorithms donde uno tendrá una complejidad temporal O(log(n)*n) y el otro de O(n * n!).

El segundo dependerá de la aleatorización para cumplir su objetivo, y se medirá el tiempo de duración en que ambos realizaron la tarea.

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2

        L = arr[:mid]
        R = arr[mid:]

        merge_sort(L)
        merge_sort(R)

        i = j = k = 0

        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1

```

```python
def is_sorted(arr):
    for i in range(len(arr) - 1):
        if arr[i] > arr[i + 1]:
            return False
    return True

def bogo_sort(arr):
    while not is_sorted(arr):
        random.shuffle(arr) # Aleatorizar arreglos (lista)
```

```python
# Lista con 8 aleatorios
arr = [random.randint(0, 100) for i in range(8)] # Aqui creamos numeros aleatorios
inicio = time.time() # Iniciamos el tiempo
merge_sort(arr)
fin = time.time() # Finalizamos el tiempo
print(fin - inicio)

inicio = time.time()
bogo_sort(arr)
fin = time.time()
print(fin - inicio)
```

# **Ejemplo 4**
Crear funciones en python, con o sin necesariamente usar los modulos importados.

```python
def fib(n:int) -> int:
  if n < 0 or type(n) == float:
    raise ValueError("El numero debe ser un entero positivo")
  if n == 0:
    return 0
  elif n == 1:
    return 1
  else:
    return fib(n-1) + fib(n-2)

fib(10)
```

```python
import graphviz

def fib_tree(n, parent=None, graph=None):
    if graph is None:
        graph = graphviz.Digraph()

    current = f"fib({n})"
    graph.node(current)

    if parent:
        graph.edge(parent, current)

    if n == 0 or n == 1:
        return graph
    else:
        fib_tree(n-1, current, graph)
        fib_tree(n-2, current, graph)

    return graph

# Crear el árbol para fib(10)
tree = fib_tree(10)
tree.render("fib_recursion_tree", format="png", cleanup=True)
```

```python
import matplotlib.pyplot as plt
from PIL import Image

img = Image.open("fib_recursion_tree.png")
plt.imshow(img)
plt.show()
```

```python
def factorial(n:int) -> int:
  if n < 0 or type(n) == float:
    raise ValueError("El numero debe ser un entero positivo")
  if n == 0:
    return 1
  else:
    return n * factorial(n-1)

print(factorial(5))
try:
  factorial(-1)
except ValueError as e:
  print(e)
```

```python
def factorial_tree(n, parent=None, graph=None):
    if graph is None:
        graph = graphviz.Digraph()

    current = f"fact({n})"
    graph.node(current)

    if parent:
        graph.edge(parent, current)

    if n == 0 or n == 1:
        return graph
    else:
        factorial_tree(n-1, current, graph)

    return graph

# Crear el árbol para fact(5)
tree = factorial_tree(5)
tree.render("factorial_recursion_tree", format="png", cleanup=True)

img = Image.open("factorial_recursion_tree.png")
plt.imshow(img)
plt.show()
```

```python
def mcd(a, b):
  if b == 0:
    return a
  else:
    return mcd(b, a % b)

mcd(12, 18)
```

```python
def mcm(a, b):
  return a * b // mcd(a, b)

mcm(12, 18)
```

```python
def probarInt(cadena: str) -> int:
  try:
    return int(cadena)
  except ValueError:
    pass
  while True:
    cadena = input('Ingrese un numero entero: ')
    try:
      return int(cadena)
    except ValueError:
      print("Necesita ser un numero entero.")
      continue

def probarFloat(cadena: str) -> float:
  try:
    return float(cadena)
  except ValueError:
    pass
  while True:
    cadena = input('Ingrese un numero flotante: ')
    try:
      return float(cadena)
    except ValueError:
      print("Necesita ser un numero flotante.")
```

```python
diccionario = {}
while True:
  menu = input("1. Ingresar paciente.\n2. Generar IMC. \n5. Salir\n")
  menu = probarInt(menu)
  if menu == 1:
    id = input("Ingrese el id del paciente: ")
    nombre = input("Ingrese el nombre del paciente: ")
    edad = probarInt(input("Ingrese la edad del paciente: "))
    genero = input("Ingrese el genero del paciente: ")
    altura = probarFloat(input("Ingrese la altura del paciente: ")) # float
    peso = probarFloat(input("Ingrese el peso del paciente: ")) # float
    if id in diccionario:
      print("El id ya existe.")
      continue
    diccionario[id] = [nombre, edad, genero, altura, peso]
  if menu == 2:
    diccionario = dict(map(lambda x: (x[0], x[1] + [x[1][4] / (x[1][3] ** 2)]), diccionario.items()))

  if menu == 3:
    print(diccionario)
  if menu == 5:
    break
```
